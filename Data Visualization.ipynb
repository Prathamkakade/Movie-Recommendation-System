```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('visual_data.csv')
```

## IMDB Rating V/S Duration

```python
temp_df_sort = df[['duration','imdb_score']].sort_values(by=['duration'])
```

```python
temp_df_sort_avg = temp_df_sort.groupby(by=["duration"]).mean()
```

```python
temp_df_sort_avg = temp_df_sort_avg.reset_index()
```

```python
#converting duration and imdb score columns to list
dur = temp_df_sort_avg['duration'].values.tolist()
rate = temp_df_sort_avg['imdb_score'].values.tolist()
```

```python
# y-axis
final_rate = []
for i in range(0,90,10):
    sm = 0
    for j in range(10):
        #print(j+i)
        sm += rate[j+i]
    final_rate.append(sm/10)
```

```python
# x-axis
dur_range = ['63-74','75-84','85-94','95-104','105-114','115-124','125-134','135-144','145-154']
```

```python
plt.plot(dur_range,final_rate)
plt.ylabel('IMDB Rating',fontsize=18)
plt.xlabel('Duration',fontsize=18)
plt.title('Variation in Average IMDB Rating with Duration of the Movie', fontsize=18)
plt.rcParams["figure.figsize"] = (20,10)
plt.grid()
plt.show()
```

## Genres V/S imdb_rating

```python
df.columns
```

```python
genre_lst = ['Adventure', 'Animation', 'Biography', 'Comedy', 'Crime', 'Documentary',
       'Drama', 'Family', 'Fantasy', 'Film-Noir', 'History', 'Horror', 'Music',
       'Musical', 'Mystery', 'News', 'Romance', 'Sci-Fi', 'Sport', 'Thriller','War', 'Western']
```

```python
len(genre_lst)
```

```python
"""Calculating average imdb rating for all genres"""

genre_rating = []
for i in genre_lst:
    avg_rate = 0
    den=0
    for j in range(4506):
        if df.loc[j,i]==1:
            avg_rate += df.loc[j,'imdb_score']
            den += 1
    genre_rating.append(round(avg_rate/den,2))
```

```python
# Figure Size
fig, ax = plt.subplots(figsize =(20, 15))
 
# Horizontal Bar Plot
ax.barh(genre_lst,genre_rating, ec="black")

# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 10)
ax.yaxis.set_tick_params(pad = 5)
 

# Add x, y gridlines
ax.grid(b = True, color ='grey',linestyle ='-.', linewidth = 0.5, alpha = 0.2)

#Show top values, i.e, in alphabetical order on y axis
ax.invert_yaxis()

# Setting x and y labels
ax.set_xlabel('Average IMDB Rating', fontsize=20)
ax.set_ylabel('Genres', fontsize=20)

# change the range of x-axis and y-axis
plt.xlim(0,9)


# Set tick font size
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(14)


# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 15, fontweight ='bold',
             color ='grey')
    
# Add Plot Title
ax.set_title('Average IMDB Rating among Genres during 1927-2016 ', loc ='left', fontsize=20)
```
