## Title Year to Int

```python
df["title_year"]=df["title_year"].astype(int)
```

## removing outliers in Duration

```python
q3=df.quantile(0.75)['duration']
q1=df.quantile(0.25)['duration']
IQR=q3-q1
```

```python
index_drop = df[(df['duration']>q3+1.5*IQR) & (df['duration']<q1-1.5*IQR) ].index
index_drop
```

```python
index_drop = df[(df['duration']>q3+1.5*IQR) | (df['duration']<q1-1.5*IQR) ].index
len(list(index_drop))
```

```python
df.drop(index_drop, inplace=True)
df
```

## One Hot encoding on genre

```python
df.reset_index(inplace=True,drop=True)
```

```python
df
```

```python
df['genres'] = df['genres'].apply(lambda x: x.split('|'))
```

```python
df_long = df.explode('genres')
```

```python
df_long
```

```python
df_one_hot_encoded = pd.concat([df, pd.get_dummies(df_long['genres'])], axis=1)
```

```python
df_one_hot_encoded=df_one_hot_encoded.groupby(df_one_hot_encoded.index).max().reset_index(drop=True)
```

```python
df_one_hot_encoded_new= pd.concat([df_one_hot_encoded, pd.get_dummies(df_one_hot_encoded['language'])], axis=1)
```

```python
df_one_hot_encoded_new= pd.concat([df_one_hot_encoded_new, pd.get_dummies(df_one_hot_encoded_new['country'])], axis=1)
```

```python
df_one_hot_encoded_new= pd.concat([df_one_hot_encoded_new, pd.get_dummies(df_one_hot_encoded_new['content_rating'])], axis=1)
```
